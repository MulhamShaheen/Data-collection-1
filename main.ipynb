# Dataset for activity based music recommendation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_dining = pd.read_csv('dining.csv', sep=',')
df_sport = pd.read_csv('sport.csv', sep=',')
df_dance = pd.read_csv('dance.csv', sep=',')
df_lounge = pd.read_csv('lounge.csv', sep=',')
```

```python
df_dining.head()
```

```python
colors = ["blue", "red", "green", "yellow"]
labels = ["dance", "sport", "lounge", "dining"]
```

```python
fig = plt.figure()

for i, frame in enumerate([df_dance, df_sport, df_lounge, df_dining]):
    plt.hist(frame['danceability'], color=colors[i], label=labels[i])

plt.legend(loc="upper left")
plt.xlabel("danceability")
plt.ylabel("count")
plt.show()
```

```python
fig = plt.figure()

for i, frame in enumerate([df_dance[:50], df_sport, df_lounge, df_dining]):
    plt.hist(frame['BPM'], color=colors[i], label=labels[i], density=1)

plt.legend(loc="upper left")
plt.xlabel("BPM")
plt.ylabel("count")
plt.show()
```

```python
fig = plt.figure()

for i, frame in enumerate([df_dance[:50], df_sport, df_lounge, df_dining]):
    plt.hist(frame['acousticness'], color=colors[i], label=labels[i])

plt.legend(loc="upper left")
plt.xlabel("Acousticness")
plt.ylabel("count")
plt.show()
```
